```python
from pandas.io.common import get_lzma_file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplts for the HD95
# Read the data from the Excel spreadsheet
Models = pd.read_excel('/content/HD95_box_updated.xlsx')


fig, axes = plt.subplots(nrows=1, ncols=2)
#Models.boxplot(column='fine', ax=axes[0], showfliers=True)
#Models.boxplot(column='gli_data', ax=axes[1], showfliers=True)

sns.boxplot(data=Models, x='subregion', y='fine', ax=axes[0])
sns.boxplot(data=Models, x='subregion', y='gli_data', ax=axes[1])

plt.title('Boxplots for the HD95 metric')



plt.show()


```

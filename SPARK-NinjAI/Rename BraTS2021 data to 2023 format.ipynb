```python
import os

# Specify the path to the main folder containing subfolders
main_folder_path = r'C:\Users\MAILAB\Desktop\ExperimentDataset\BraTS2021_TrainingData'

# List all subdirectories in the main folder
subfolder_names = os.listdir(main_folder_path)

# Initialize counters for uniform and non-uniform subfolders
uniform_subfolders = 0
non_uniform_subfolders = 0

# Loop through each subfolder
for subfolder_name in subfolder_names:
    if os.path.isdir(os.path.join(main_folder_path, subfolder_name)):
        subfolder_path = os.path.join(main_folder_path, subfolder_name)
        file_names = os.listdir(subfolder_path)
        
        subfolder_prefix = subfolder_name + '-'
        subfolder_uniform = True
        
        # Loop through each file in the subfolder
        for file_name in file_names:
            if file_name.endswith('-seg.nii.gz') or file_name.endswith('-t1n.nii.gz') or file_name.endswith('-t1c.nii.gz') or file_name.endswith('-t2w.nii.gz') or file_name.endswith('-t2f.nii.gz'):
                if not file_name.startswith(subfolder_prefix):
                    subfolder_uniform = False
                    break
        
        if subfolder_uniform:
            uniform_subfolders += 1
            print(f"Subfolder '{subfolder_name}' and its contents are uniform.")
        else:
            non_uniform_subfolders += 1
            print(f"Subfolder '{subfolder_name}' and its contents are NOT uniform.")

print(f"Total uniform subfolders: {uniform_subfolders}")
print(f"Total non-uniform subfolders: {non_uniform_subfolders}")
```
